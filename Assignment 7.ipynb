```python
#  class to control a light

class SmartLight:

    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # light starts as OFF

    def turn_on(self):
        self.status = "ON"
        print(self.name, "is", self.status)

    def turn_off(self):
        self.status = "OFF"
        print(self.name, "is", self.status)


light1 = SmartLight("Bedroom Light")
light1.turn_on()
```

```python
# storing employee information

class Employee:

    def __init__(self, name, emp_id, dept):
        self.name = name
        self.emp_id = emp_id
        self.dept = dept

    def show_details(self):
        print("Employee ID Card")
        print("Name:", self.name)
        print("ID:", self.emp_id)
        print("Department:", self.dept)


emp1 = Employee("Rani", "EMP102", "IT")
emp1.show_details()
```

```python
# contact storage

class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c1 = Contact("Rani", "9876543210")
c1.show_contact()
```

```python
# # storing movie info

class Movie:

    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def show_movie(self):
        print("Movie:", self.title)
        print("Rating:", self.rating, "/ 5")


m1 = Movie("Joker", 4.8)
m1.show_movie()

class Contact:

    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def show_contact(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)


c1 = Contact("Rani", "9876543210")
c1.show_contact()
```

```python
# class for product price

class Product:

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def price_tag(self):
        print("Product:", self.name)
        print("Price: ₹", self.price)


p1 = Product("Headphones", 2499)
p1.price_tag()
```

```python
# delivery information

class Delivery:

    def __init__(self, customer, address):
        self.customer = customer
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.customer)
        print("Address:", self.address)


d1 = Delivery("Rani", "Hyderabad")
d1.display()
```
